```python
import pandas as pd
from pycaret.regression import *
```

```python
data = pd.read_csv("Fuel_cell_performance_data-Full.csv")

target_column = 'Target1'

data = data.drop(columns=['Target2', 'Target3', 'Target4', 'Target5'])

print(data.columns)
```

```python
reg_setup = setup(
    data=data, 
    target=target_column, 
    train_size=0.7, 
    normalize=True,
    transformation=True
)
```

```python
top_models = compare_models(n_select=4)
```

####  Extra Tree Regreesor Bayesian Ridge, Linear Regression & Ridge Regression are best on training data so we will train and compare results of unseen data on each of them

```python
final_models = [finalize_model(model) for model in top_models]

predictions = [predict_model(model) for model in final_models]
```

### All 4 models were very close on accuracy (R2) on training set. But Extra Trees Regressor but extremely well on Test set. So this is best model and we will save it  

```python
best_model = finalize_model(top_models[0])
best_model
```

```python
save_model(best_model, 'best_regression_model')
```
